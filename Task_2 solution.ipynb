```python
import yfinance as yf
from datetime import datetime, timedelta, date

def fetch_store_15min_data(stock_symbol, num_days):
    # Create a Ticker object for the specified stock
    stock = yf.Ticker(stock_symbol)

    # Get today's date
    today = date.today()

    # Loop through the specified number of days
    for day in range(num_days):
        current_date = today - timedelta(days=day)
        start_time = datetime.combine(current_date, datetime.strptime("11:15", "%H:%M").time())
        end_time = datetime.combine(current_date, datetime.strptime("14:15", "%H:%M").time())

        # Fetch 15-minute interval data for the current day
        data = stock.history(interval='15m', start=start_time, end=end_time)

        # Print or store the data as needed
        print(f"15-minute candle data for {stock_symbol} on {current_date}:")
        print(data)

if __name__ == "__main__":
    # Define parameters
    icici_symbol = "ICICIBANK.NS"
    num_days = 7  # Specify the number of days

    # Call the function to fetch and store 15-minute candle data
    fetch_store_15min_data(icici_symbol, num_days)
```
